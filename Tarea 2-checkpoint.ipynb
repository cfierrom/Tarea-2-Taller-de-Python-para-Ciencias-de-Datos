Separamos los Dataframe por tipo de vino, para obtener los promedios de los datos faltantes y remplazar estos datos. Luego volveremos a concatenar estos DF


   

```python
df_white = df[df['Type-Wine']=='White']
df_red = df[df['Type-Wine']=='Red']
```

```python
df_white.isnull().sum(axis = 0)
```

```python
df_red.isnull().sum(axis = 0)
```

```python

prom_sug = df_white["residual sugar"].astype("float").mean(axis=0)
prom_density = df_white["density"].astype("float").mean(axis=0)
prom_alcohol = df_white["alcohol"].astype("float").mean(axis=0)

print("Promedio de azucar residual:", prom_sug)
print("Promedio de la densidad:", prom_density)
print("Promedio del alcohol:", prom_alcohol)

df_white["residual sugar"].replace(np.nan, prom_sug, inplace=True)
df_white["density"].replace(np.nan, prom_density, inplace=True)
df_white["alcohol"].replace(np.nan, prom_alcohol, inplace=True)

```

```python
prom_sug = df_red["residual sugar"].astype("float").mean(axis=0)
prom_density = df_red["density"].astype("float").mean(axis=0)
prom_alcohol = df_red["alcohol"].astype("float").mean(axis=0)

print("Promedio de azucar residual:", prom_sug)
print("Promedio de la densidad:", prom_density)
print("Promedio del alcohol:", prom_alcohol)

df_red["residual sugar"].replace(np.nan, prom_sug, inplace=True)
df_red["density"].replace(np.nan, prom_density, inplace=True)
df_red["alcohol"].replace(np.nan, prom_alcohol, inplace=True)
```

```python
df=pd.concat([df_red, df_white]).reset_index(drop=True)
df
```

```python
import matplotlib.pyplot as plt
#from matplotlib import pyplot

plt.hist(df["quality"])

# setear las etiquetas para los ejes x e y más un título
plt.xlabel("quality")
plt.ylabel("cantidad")
plt.title("Contenedores quality")
```

```python
contenedores = np.linspace(min(df["quality"]), max(df["quality"]), 4)
contenedores
nombres_de_grupo = ['Mala', 'Media', 'Alta']
df['quality classification'] = pd.cut(df['quality'], contenedores, labels=nombres_de_grupo, include_lowest=True )
df[['quality','quality classification']].head(20)
df
```

Grafico de barras segun calidad (Mala, Media, Alta) de los vinos(vino blanco y vino rojo)

```python

plt.bar(nombres_de_grupo, df["quality classification"].value_counts())

# setear las etiquetas para los ejes x e y más un título
plt.xlabel("quality")
plt.ylabel("cantidad")
plt.title("Quality Wine")
```

Primero separamos los DF en vino blanco y vino rojo para graficar por separado, ademas se crea una funcion para agregar las cantidad correspondientes a cada barra(mala,media,alta), ya que sin esto es dificil aprecia los valores que estos puedan poseer.

```python
df_white = df[df['Type-Wine']=='White']
df_red = df[df['Type-Wine']=='Red']

def agregar_valor_columna(x_list,y_list):
    for i in range(0, len(x_list)):
        plt.text(i,y_list[i-1]/2,y_list[i-1], ha="center")


plt.bar(nombres_de_grupo,df_white["quality classification"].value_counts() )
valores=[1640,180,3078]
agregar_valor_columna(nombres_de_grupo,valores)
plt.title("Quality Wine")
plt.xlabel("Quality")
plt.ylabel("Cantidad")
plt.show()
```

```python
plt.bar(nombres_de_grupo,df_red["quality classification"].value_counts() )
valores=[837,18,744]
agregar_valor_columna(nombres_de_grupo,valores)
plt.title("Quality Wine")
plt.xlabel("Quality")
plt.ylabel("Cantidad")
plt.show()
df_red["quality classification"].value_counts()
```

```python

plt.bar(nombres_de_grupo, df_white["quality classification"].value_counts(), width = 0.4)
plt.bar(nombres_de_grupo, df_red["quality classification"].value_counts(), width = 0.4)
# setear las etiquetas para los ejes x e y más un título
plt.xlabel("Quality")
plt.ylabel("cantidad")
plt.title("Quality Wine")
plt.legend(('White Wine', 'Red Wine'))
```
