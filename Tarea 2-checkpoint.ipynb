Basado en el resumen anterior, cada columna tiene 4898 filas de datos y 12 columnas. En 2 columnas podemos observar datos faltantes.
 
1. En la columna residual sugar, 25 datos faltantes.
2. En la columna density, 147 datos faltantes.
3. En la columna alcohol, 40 datos faltantes.


Separamos los Dataframe por tipo de vino, para obtener los promedios de los datos faltantes y remplazar estos datos.

   

```python
df_white = df[df['Type-Wine']=='White']
df_red = df[df['Type-Wine']=='Red']
```

```python
df_white.isnull().sum(axis = 0)
```

```python
df_red.isnull().sum(axis = 0)
```

```python

prom_sug = df_white["residual sugar"].astype("float").mean(axis=0)
prom_density = df_white["density"].astype("float").mean(axis=0)
prom_alcohol = df_white["alcohol"].astype("float").mean(axis=0)

print("Promedio de azucar residual:", prom_sug)
print("Promedio de la densidad:", prom_density)
print("Promedio del alcohol:", prom_alcohol)

df_white["residual sugar"].replace(np.nan, prom_sug, inplace=True)
df_white["density"].replace(np.nan, prom_density, inplace=True)
df_white["alcohol"].replace(np.nan, prom_alcohol, inplace=True)

```

```python
prom_sug = df_red["residual sugar"].astype("float").mean(axis=0)
prom_density = df_red["density"].astype("float").mean(axis=0)
prom_alcohol = df_red["alcohol"].astype("float").mean(axis=0)

print("Promedio de azucar residual:", prom_sug)
print("Promedio de la densidad:", prom_density)
print("Promedio del alcohol:", prom_alcohol)

df_red["residual sugar"].replace(np.nan, prom_sug, inplace=True)
df_red["density"].replace(np.nan, prom_density, inplace=True)
df_red["alcohol"].replace(np.nan, prom_alcohol, inplace=True)
```
